#box plot for each and every numerical column

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
box = plt.boxplot(df['Avg Price'])
plt.show
```

```python
#extracting outliers

outliers = []
for line in box['fliers']:
    outliers.extend(line.get_ydata())  # use get_ydata() if vert=True
print(outliers)
```
