```python
df1 = df.drop('ID#', axis = 1)
```

```python
df1.columns
```

# Model Building

```python
X = df1.drop(columns=['Award?'])  # Features
y = df1['Award?']  # Target Variable (0 or 1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
# Apply Agglomerative Clustering
AC = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')  # 2 clusters since target is binary
clusters_ac = AC.fit_predict(X_scaled)
```

```python
# Add cluster labels to DataFrame for analysis
df1['Cluster_ac'] = clusters_ac
```

```python
df1.head()
```

```python
from sklearn.metrics import silhouette_score
```

```python
sil_score_Hierarchy = silhouette_score(X_scaled, clusters_ac)
```

```python
print(f"Silhouette Score(Agglomerative): {sil_score_Hierarchy:.3f}")
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
cr = classification_report(df1['Award?'], df1['Cluster_ac'])
```

```python
print(cr)
```

# KMeans Clustering

```python
from sklearn.cluster import KMeans 
```

```python
kmeans = KMeans(n_clusters=2, random_state=42, n_init=6)
clusters_kmeans = kmeans.fit_predict(X_scaled)
```

```python
cluster_labels = np.where(clusters_kmeans == pd.Series(clusters_kmeans)[y == 1].mode()[0], 1, 0)
```

```python
df1['Cluster_km'] = clusters_kmeans
```

```python
df1.head()
```

```python
sil_score_kmeans = silhouette_score(X_scaled, clusters_kmeans)
print(f"Silhouette Score(kmeans): {sil_score_kmeans:.3f}")
```

```python
cr_km = classification_report(df1['Award?'], df1['Cluster_km'])
```

```python
print(cr_km)
```

# DBScan Clustering

```python
from sklearn.cluster import DBSCAN
```

```python
cluster_dbscan = DBSCAN(eps=0.8, min_samples=12)
cluster_dbscan.fit(X_scaled)
```

```python
cluster_dbscan.labels_
```

```python
cluster_Dbs=pd.DataFrame(cluster_dbscan.labels_,columns=['cluster_Dbs'])
```

```python
df1['cluster_Dbs'] = cluster_dbscan.labels_
```

```python
df1
```

```python
sil_score_dbs = silhouette_score(X_scaled, cluster_Dbs)
print(f"Silhouette Score(dbscan): {sil_score_dbs:.3f}")
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

# Visualization of Clustering

```python
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters_ac, palette="coolwarm")
plt.title("Agglomerative Clustering")
```

```python
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters_kmeans, palette="viridis")
plt.title("K-Means Clustering")
```

```python
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=cluster_dbscan, palette="viridis")
plt.title("DBScan Clustering")
```

Among three clustering methods Kmeans clustering methods gives most precise and high silhouette score cluster.
DBScan method is used for large data where noise is considered as -1, here in this dataset we cant cluster -1 point due to its 
binary classification 
silhouette scores Kmeans > DBScan > Agglomerative
