Q) Summarize the key insights and patterns discovered through your exploratory analysis.

Key Correlations
Strong Positive Correlations:
ASTV and NSP (0.457):

This indicates that as the "Percentage of Time with Abnormal Short-Term Variability" (ASTV) increases, the probability of a specific class in the "NSP" (normal, suspect, or pathological) label also increases. ASTV might be a key predictor of fetal health status.
MSTV and Width (0.661):

A strong relationship exists between the "Mean Value of Short-Term Variability" (MSTV) and "Width," which may suggest these variables together influence fetal conditions or health trends.
ALTV and ASTV (0.414):

"Percentage of Time with Abnormal Long-Term Variability" (ALTV) and ASTV are positively correlated. This could mean that variability in short and long-term measures tends to align in abnormal fetal conditions.
DP and NSP (0.294):

"Decelerations Prolonged" (DP) has a moderately positive correlation with NSP, implying that prolonged decelerations could indicate abnormal fetal conditions.
Strong Negative Correlations:
MSTV and ASTV (-0.495):

There's a negative correlation between "Mean Value of Short-Term Variability" (MSTV) and "Percentage of Time with Abnormal Short-Term Variability" (ASTV). This suggests that higher variability in the abnormal range may reduce the mean variability overall.
ALTV and MSTV (-0.609):

A strong negative correlation indicates that abnormal long-term variability (ALTV) reduces the mean short-term variability (MSTV). These metrics might inversely reflect different aspects of fetal heart health.
AC and ALTV (-0.428):

Accelerations (AC) are negatively correlated with "Percentage of Time with Abnormal Long-Term Variability" (ALTV). This suggests that fetal movements with higher accelerations may lead to reduced long-term abnormal variability.
Potential Insights:
ASTV and ALTV: Both seem critical to determining NSP, as they show significant correlations with this target variable.
MSTV and Width: Their strong relationship might indicate mechanical or physiological factors influencing fetal variability.
Prolonged Decelerations (DP): Could be a warning sign for abnormal conditions given its moderate positive correlation with NSP.
