Discuss the pros and cons of One-Hot Encoding and Label Encoding

One-Hot encoding:-
Pros:
No ordinal relationship is assumed between categories
Works well with algorithms that don’t handle categorical data natively.
Prevents unintended relationships between values.

Cons:
Increases dimensionality.
Can cause performance issues on high-cardinality columns.

Label Encoding:-
Pros:
Simple and memory efficient
Does not increase dimensionality
Useful for ordinal data where category order matters (e.g., low < medium < high)

Cons:
Implies an ordinal relationship even if one doesn’t exist — can mislead algorithms
Can cause models to learn wrong patterns due to the numeric nature of labels
