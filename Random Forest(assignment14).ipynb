# Model Buliding 

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
df
```

```python
X = df.iloc[:,0:9]
Y = df['Type']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Random Forest Classification
kfold = KFold(n_splits=10, random_state=7, shuffle=True)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
# Bagging Classifier
bag_model = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=100, random_state=42)
bag_model.fit(X_train, y_train)
bag_pred = bag_model.predict(X_test)
```

```python
# Boosting Classifier
boost_model = AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1), n_estimators=100, random_state=42)
boost_model.fit(X_train, y_train)
boost_pred = boost_model.predict(X_test)
```

# evaluating the model

```python
def evaluate_model(name, y_true, y_pred):
    print(f"\n{name} Performance:")
    print("Accuracy :", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred, average='weighted'))
    print("Recall   :", recall_score(y_true, y_pred, average='weighted'))
    print("F1 Score :", f1_score(y_true, y_pred, average='weighted'))
```

```python
evaluate_model("Random Forest", y_test, rf_pred)
```

```python
evaluate_model("Bagging", y_test, bag_pred)
```

```python
evaluate_model("Boosting (AdaBoost)", y_test, boost_pred)
```

# Q&A

Explain Bagging and Boosting methods. How is it different from each other.

Bagging and Boosting are both ensemble methods used to improve the performance of machine learning models by combining multiple base learners. Bagging builds several independent models  on different random subsets of the training data , and then aggregates their predictions to reduce variance and prevent overfitting.
Boosting builds models sequentially, where each new model focuses on correcting the errors made by the previous ones, thereby reducing bias and improving accuracy .
The main difference is : bagging trains in parallel and aims to reduce variance, while boosting trains sequentially and focuses on reducing bias
